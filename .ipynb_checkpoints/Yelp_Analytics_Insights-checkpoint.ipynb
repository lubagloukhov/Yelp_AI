### Cluster:
1. Assign cluster membership to 1 of 3 clusters based on lat/long of each user's reviews
2. Identify the cluster with the most reviews for each user as "homebase"

```python
df100_group = df100.groupby(['User - Id'])
```

```python
km = KMeans(3) # initialize

def fitpred(X):
    km.fit(X[['Latitude','Longitude']])
    X['cluster'] = km.predict(X[['Latitude','Longitude']])
    X['numrevs'] = X.groupby(['cluster'])['Latitude'].transform('count')
    X['homebase'] = X['numrevs']==np.max(X['numrevs'])
    
    X['cc_LatLon'] = [km.cluster_centers_[x] for x in km.predict(X[['Latitude','Longitude']])]
    X['cc_Lat'] = [x[0] for x in X['cc_LatLon']]
    X['cc_Lon'] = [x[1] for x in X['cc_LatLon']]

    return X
```

```python
df100 = df100_group.apply(fitpred).sort_values('User - Id')
```

### Response Variable: "Business - Price Range"

People are more likely to review a higher priced restaurant away from homebase? When away from homebase, reviews are for higher priced locations but with higher std.

```python
print df100.groupby('homebase')["Business - Price Range"].agg([len, np.mean, np.std]).round(2)
g = sns.FacetGrid(df100, col="homebase")
g.map(sns.countplot, "Business - Price Range", palette="Greens")
```

```python
sns.boxplot(x="homebase", y="Business - Price Range", data=df100)
plt.title("Distribution of Business Price Range")
```

We need to take into consideration individual variation. 
We can do this by subtracting the individual average. 
Individuals tend do rate cheaper venues when they are in their homebase.


```python
df100_bpr = df100[['User - Id','homebase','Business - Price Range']]
df100_bpr['BPR_useravg'] = df100_bpr.groupby('User - Id')['Business - Price Range'].transform('mean')
df100_bpr['BPR_demeaned'] = df100_bpr['Business - Price Range'] - df100_bpr['BPR_useravg']

print df100_bpr.groupby('homebase')["BPR_demeaned"].agg([len, np.mean, np.std]).round(2)
sns.boxplot(x="homebase", y="BPR_demeaned", data=df100_bpr)

plt.xlabel('homebase')
plt.ylabel('User-Demeaned Business - Price Range')
plt.title("Distribution of User-Demeaned Business Price Range", y=1.03)
```

Are these differences statistically significant? Can we make conclusions abou the population of users from the observations on this sample of data?

```python
# Compute Two sample Welch's t-test statistic & p-value
a = df100_bpr["BPR_demeaned"][df100_bpr['homebase']==True]
b = df100_bpr["BPR_demeaned"][df100_bpr['homebase']==False]

abar = a.mean()
avar = a.var(ddof=1)
na = a.size
adof = na - 1

bbar = b.mean()
bvar = b.var(ddof=1)
nb = b.size
bdof = nb - 1

tf = (abar - bbar) / np.sqrt(avar/na + bvar/nb)
dof = (avar/na + bvar/nb)**2 / (avar**2/(na**2*adof) + bvar**2/(nb**2*bdof))

pf_Ha_dnon0 = 2*stdtr(dof, -np.abs(tf)) 

print tf, pf_Ha_dnon0 
```

From the test-statistic we can compute the 99% confidence interval for the differenece in means

```python
t_star = stats.t.isf([0.01/2], [dof])
lb = (abar - bbar) - t_star*np.sqrt(avar/na + bvar/nb)
ub = (abar - bbar) + t_star*np.sqrt(avar/na + bvar/nb)
print lb, ub
```

The  exceptionally small P-value and non-zero-inclusive confience interval would lead us to reject the null hypothesis of no difference in means in favor of the alternate hypothesis of a non-zero difference in means. 

There is a statistical difference between the 'Business - Price Range' of homebase and non-homebase reviews, when controlling for individual differences.

# Aside/Consideration:
## "Review - Stars"

People are more likel to leave higher reviews when away from homebase? When away from Homebase, reviews are ever so slightly higher but with higher cov.

```python
print df100.groupby('homebase')["Review - Stars"].agg([len, np.mean, np.std])
g = sns.FacetGrid(df100, col="homebase")
g.map(sns.countplot, "Review - Stars");
```

```python
sns.boxplot(x="homebase", y="Review - Stars", data=df100)
```

We need to take into consideration individual variation by subtracting the individual average. Individuals tend do rate lower when they are in their homebase.

```python
df100_rs = df100[['User - Id','homebase',"Review - Stars"]]
df100_rs['RS_useravg'] = df100_rs.groupby('User - Id')["Review - Stars"].transform('mean')
df100_rs['RS_demeaned'] = df100_rs["Review - Stars"] - df100_rs['RS_useravg']

print df100_rs.groupby('homebase')["RS_demeaned"].agg([len, np.mean, np.std])
sns.boxplot(x="homebase", y="RS_demeaned", data=df100_rs)
```

Are these differences statistically significant?


```python
a = df100_rs["RS_demeaned"][df100_rs['homebase']==True]
b = df100_rs["RS_demeaned"][df100_rs['homebase']==False]

abar = a.mean()
avar = a.var(ddof=1)
na = a.size
adof = na - 1

bbar = b.mean()
bvar = b.var(ddof=1)
nb = b.size
bdof = nb - 1

tf = (abar - bbar) / np.sqrt(avar/na + bvar/nb)
dof = (avar/na + bvar/nb)**2 / (avar**2/(na**2*adof) + bvar**2/(nb**2*bdof))
pf_Ha_dnon0 = 2*stdtr(dof, -np.abs(tf)) 
print tf, pf_Ha_dnon0
```

In this example, the two-tailed p-value suggests rejecting the null hypothesis of no difference in favor of the laternate hypothesis of a non-zero difference at a confidence level of .05 but not a more conservative one of .01.

### Does the above change if we demean by User - Average Stars?

Demeaning by the greater population mean (User - Average Stars) as opposed to the data sample mean (average within the data we have at hand).


```python

df100_rs2 = df100[['User - Id','homebase',"Review - Stars", "User - Average Stars"]]
df100_rs2['RS_demeaned'] = df100_rs2["Review - Stars"] - df100_rs2["User - Average Stars"]

print df100_rs2.groupby('homebase')["RS_demeaned"].agg([len, np.mean, np.std])
sns.boxplot(x="homebase", y="RS_demeaned", data=df100_rs2, palette="PRGn")
```

Are these differences statistically significant?

```python
a = df100_rs2["RS_demeaned"][df100_rs2['homebase']==True]
b = df100_rs2["RS_demeaned"][df100_rs2['homebase']==False]

abar = a.mean()
avar = a.var(ddof=1)
na = a.size
adof = na - 1

bbar = b.mean()
bvar = b.var(ddof=1)
nb = b.size
bdof = nb - 1

tf = (abar - bbar) / np.sqrt(avar/na + bvar/nb)
dof = (avar/na + bvar/nb)**2 / (avar**2/(na**2*adof) + bvar**2/(nb**2*bdof))
pf_Ha_dnon0 = 2*stdtr(dof, -np.abs(tf)) 

print tf, pf_Ha_dnon0
```

The two-tailed p-value suggests failing to reject the null hypothesis of no difference at a confidence level of .05. These results are even weaker than the sample demeaning case above.
